```
# Import Python Libraries

import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```
# Read csv file

df_partA = pd.read_csv('https://vincentarelbundock.github.io/Rdatasets/csv/datasets/mtcars.csv')
```

###2. Display the first five rows of the data.

```
# List first 5 row

df_partA.head(n=5)
```

###3. For each category in the cyl column, calculate the average mpg for all cars with that cyl value.

```
df_partA.groupby('cyl')[['mpg']].mean()
```

###4. Create a histogram using the mpg column.

```
# Histograms
df_partA['mpg'].hist()
plt.xlabel('mpg')
plt.ylabel('Count')
```

###5. Choose two variables in the data and create a scatterplot.


```
# Scatterplots

plt.scatter(df_partA['hp'], df_partA['qsec'])
plt.title("Horsepower and acceleration performance")
plt.xlabel("horsepower")
plt.ylabel("1/4 mile time")

#try to figure out what is "qsec" and get the idea from: https://rstudio-pubs-static.s3.amazonaws.com/61800_faea93548c6b49cc91cd0c5ef5059894.html. hence making the scatterplot to observe the relationship between hp and qsec
```
